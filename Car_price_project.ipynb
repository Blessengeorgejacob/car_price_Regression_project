```python
dummy=pd.get_dummies(car[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
dummy
```

```python
car1=car1.drop(['Location','Fuel_Type','Transmission','Owner_Type','Name','New_Price','Unnamed: 0'],axis=1)
car1
```

```python
car1.isna().sum()
```

```python
#String Replace
car1['Mileage']=car1['Mileage'].str.replace('km/kg','')
car1['Mileage']=car1['Mileage'].str.replace('kmpl','')
car1
```

```python
car1['Engine']=car1['Engine'].str.replace('CC','')
car1
```

```python
car1['Power']=car1['Power'].str.replace('bhp','')
car1
```

```python
#null
car1['Mileage']=car1['Mileage'].str.replace('null','0')
car1['Engine']=car1['Engine'].str.replace('null','0')
car1['Power']=car1['Power'].str.replace('null','0')
car1
```

```python
car1.dtypes
```

```python
car1['Engine']=car1['Engine'].astype(float)
car1['Power']=car1['Power'].astype(float)
car1['Mileage']=car1['Mileage'].astype(float)
car1
```

```python
car1.dtypes
```

```python
car1.isna().sum()
```

```python
import numpy as np
car1.loc[car1.Engine==0,'Engine']=np.NAN  #missing_data
```

```python
car1.loc[car1.Power==0,'Power']=np.NAN
```

```python
car1.loc[car1.Mileage==0,'Mileage']=np.NAN
```

```python
car1.isna().sum()
```

```python
car1['Engine']=car1['Engine'].fillna(car1['Engine'].mean())
car1['Power']=car1['Power'].fillna(car1['Power'].mean())
car1['Mileage']=car1['Mileage'].fillna(car1['Mileage'].mean())
car1['Seats']=car1['Seats'].fillna(car1['Seats'].mode()[0])
```
